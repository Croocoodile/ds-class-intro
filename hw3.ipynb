```python
emp = pd.read_csv('data/employee.csv')
```

```python
emp.head()
```

```python
df_3 = emp.groupby(['race', 'gender']).size().reset_index(name = 'counts')
total = len(emp.dropna())
df_3['ratio'] = df_3['counts'] / total
df_3.head()
```
